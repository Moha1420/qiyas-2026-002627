```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.parse import urljoin

old_archive_news = []
new_news=[]
# Archive URLs
oldtags = [
    "https://aigaforum.com/Archive/ArchiveJan2021.htm",
    "https://aigaforum.com/Archive/ArchiveFeb2021.htm",
    "https://aigaforum.com/Archive/ArchiveMar2021.htm"
]

headers =[
"SJT","OMNA","Mekete","UK",  "Abiy Aggression News","Home","አማርኛ/ትግርኛ","Tigrai TV", 
    "DW TV"," Reyot",
    "TMH",    "Assena ","AWLO" ,"EthF",   "Archive"]

for news_url in oldtags:

    try:
        response = requests.get(news_url, timeout=10)

        if response.status_code == 200:

            soup = BeautifulSoup(response.text, "html.parser")

            print(f"Fetching: {news_url}")

            for link in soup.find_all("a"):

                title = link.get_text(strip=True)
                href = link.get("href")

                if title and href:

                    # Convert relative URL to absolute URL
                    full_url = urljoin(news_url, href)

                    old_archive_news.append({
                        "title": title,
                        "url": full_url,
                        "source_archive": news_url
                    })

    except Exception as e:
        print(f"Error fetching {news_url}: {e}")

# ----------------------------------
# Create DataFrame
# ----------------------------------
df = pd.DataFrame(old_archive_news)

# Remove duplicates
df.drop_duplicates(subset=["title", "url"], inplace=True)

# Add useful columns
df["title_length"] = df["title"].str.len()
df["url_length"] = df["url"].str.len()


# Save CSV File

csv_file = "Aiga_csv.csv"

df.to_csv(csv_file, index=False, encoding="utf-8-sig")

print(f"\nSaved {len(df)} records to {csv_file}")


# Read CSV Again

news_df = pd.read_csv(csv_file)

print("\nFirst 10 Records:")
print(news_df.head(10))
```
