```python
import requests
from bs4 import BeautifulSoup

url = "https://t.me/s/tikvahethiopia"

html = requests.get(url).text

soup = BeautifulSoup(html, "html.parser")

posts = soup.find_all("div", class_="tgme_widget_message_text")

for post in posts:
    print(post.get_text())
    print("=" * 50)
```
